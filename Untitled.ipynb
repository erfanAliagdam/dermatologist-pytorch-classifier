```python
import torch
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, Dataset
import numpy as np
import torch.nn as nn
import matplotlib.pyplot as plt
from torch import optim
import PIL
import warnings
import pyprind

warnings.filterwarnings("ignore")
#dir(transforms)
```

```python
batch_size = 5
transform  = transforms.Compose([transforms.Resize((255, 255)),
                                 transforms.CenterCrop(224),
                                 transforms.ToTensor(),
                                 transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                      std=[0.229, 0.224, 0.225])])
train_data  = datasets.ImageFolder(root="./data/train/", transform=transform)
test_data   = datasets.ImageFolder(root="./data/test/",  transform=transform)
valid_data  = datasets.ImageFolder(root="./data/valid/", transform=transform)

train_loader = DataLoader(train_data, batch_size = batch_size, shuffle=True, num_workers=4)
test_loader  = DataLoader(test_data,  batch_size = batch_size, shuffle=True, num_workers=4)
valid_loader = DataLoader(valid_data, batch_size = batch_size, shuffle=True, num_workers=4)
print(len(train_data.classes))
```

```python
classes        = train_data.classes
dataIter       = iter(train_loader)
images, labels = dataIter.next()
images = images.numpy()
fig = plt.figure(figsize = (20, 4))
for idx in np.arange(batch_size):
    ax = fig.add_subplot(2, 10, idx+1, xticks=[], yticks=[])
    img = np.transpose(images[idx], (1, 2, 0))
    img = img * 0.229 + 0.485
    plt.imshow(img)
    ax.set_title(classes[labels[idx]])
```

```python
model = models.resnet50(pretrained=True)
```

```python
print(model)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
for param in model.parameters():
    param.require_grad = False

model.fc = nn.Sequential(nn.Linear(2048, 512),
                        nn.ReLU(),
                        nn.Dropout(0.2),
                        nn.Linear(512, 64),
                        nn.ReLU(),
                        nn.Dropout(0.2),
                        nn.Linear(64, 3))
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.fc.parameters(), lr=0.003)
model.to(device)
print(device)
print(torch.cuda.is_available())
```

```python
def train(model, epochs, batch_size, save_path):
    print("training in process...")
    loss_val_min = np.Inf

    for epoch in range(epochs):
        loss_var = 0.0
        loss_val = 0.0
        model.train()
        for data, target in train_loader:
            data, target = data.to(device), target.to(device)
            optimizer.zero_grad()
            out = model.forward(data)
            loss = criterion(out, target)
            loss.backward()
            optimizer.step()
            loss_var += loss.item()
        model.eval()
        for data, target in valid_loader:
            data, target = data.to(device), target.to(device)
            out = model(data)
            loss = criterion(out, target)
            loss_val += loss.item()
            
        loss_var = loss_var/batch_size
        loss_val = loss_val/batch_size
        print(f" in epoch : {epoch} train loss is:{loss_var} and valid loss is :{loss_val}")
        if loss_val<= loss_val_min:
            print("saving model")
            torch.save(model.state_dict(), save_path)
            loss_val_min = loss_val
        print(f" epoch {epoch} finished.")
    print(" ________traigning __finished________________________")  
    return model
            
            
          
```

```python
model_trained = train(model, 20, batch_size, "modelRESNET50-5batch---2.pt")
```
